## Build out the dataset2 ##
import dataset2_crowdtruth_17k.csv (which is already deduped),

```python
df2 = pd.read_csv('data/dataset2_crowdtruth_17k.csv')
```

```python
# add a new column for not_funny label
df2["not_funny"] = abs(df2["funny"] -1)
```

```python
df2[:10]
```

```python
print("Total number of rows:",len(df2))
print("Number of rows marked funny:",len(df2[(df2['funny'] == 1)]))
```

```python
# use .casefold() if this doesn't work well?
df2['text'] = df2['text'].map(lambda x: x.lower())
```

```python
df2[:5]
```

```python
# make a df for just the non-funny documents

not_funny = df2[(df2['funny'] == 0)]
```

```python
len(not_funny)
```

```python
not_funny.head()
```

```python
not_funny_sample = not_funny.sample(n=5434)
```

```python
not_funny_sample.head()
```

```python
len(not_funny_sample)
```

```python
df_balanced = pd.concat([not_funny_sample, df2[df2["funny"] == 1]])
```

```python
df_balanced.shape
```

```python
df_balanced.head()
```

```python
len(df_balanced[df_balanced["funny"] == 1])
```

```python
print(len(df_balanced[df_balanced["funny"] == 1]),"funny-labeled jokes comprise","{:.1f}".format(len(df_balanced[df_balanced["funny"] == 1])/len(df_balanced)*100),"% of the",len(df_balanced),"row dataset2_crowdtruth_working.csv")
```

```python
df_balanced.to_csv('working_data/dataset2_crowdtruth_working.csv', index=False)
```
